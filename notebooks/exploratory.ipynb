```python
import pandas as pd
df = pd.read_csv("../data/processed/merged_panel.csv")
df.head()
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.scatter(df['Kharif_Rain_mm'], df['Yield_t_ha'], alpha=0.6)
plt.xlabel("Kharif Rainfall (mm)")
plt.ylabel("Yield (t/ha)")
plt.title("Rainfall vs Rice Yield (India 2000-2019)")
plt.show()
```

```python
df[['Kharif_Rain_mm', 'Yield_t_ha']].corr()
```

```python
import seaborn as sns
sns.lmplot(x='Kharif_Rain_mm', y='Yield_t_ha', hue='State', data=df, height=5, aspect=1.5)
```

```python
state = "Assam"
subset = df[df['State'] == state]
subset.plot(x='Year', y=['Yield_t_ha', 'Kharif_Rain_mm'], figsize=(7,4))
```
